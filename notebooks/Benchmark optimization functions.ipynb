```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

# Benchmark optimization functions

These functions were each used to analyze schemata patterns

## Booth function

**Equation**

$f(x,y)=\left(x+2y-7\right)^{2}+\left(2x+y-5\right)^{2}$

**Global optimum**

$f(1,3)=0$

**Constraints**

$-10\leq x,y\leq 10$


```python
# Define the Booth function
def booth_function(x, y):
    return (x + 2*y - 7)**2 + (2*x + y - 5)**2
```

```python
# Generate data points for visualization
x_values = np.linspace(-10, 10, 100)
y_values = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x_values, y_values)
Z = booth_function(X, Y)

# Highlight the point (1, 3) with a red dot
highlight_x = 1
highlight_y = 3
highlight_z = booth_function(highlight_x, highlight_y)

# Create a 2D and 3D subplot side by side
fig = plt.figure(figsize=(10,6))

# Plot the 2D contour plot
ax = fig.add_subplot(1,2,1)
contour = ax.contourf(X, Y, Z, cmap='viridis', levels=30)
ax.scatter([highlight_x], [highlight_y], color='red', label='f(1, 3)=0')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Booth Function Contour Plot with Shading')
ax.legend()
fig.colorbar(contour, ax=ax)

# # Plot the 3D surface plot with only 3D axes
ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.scatter([highlight_x], [highlight_y], [highlight_z], color='red', label='f(1, 3)=0')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Booth Function Value')

# Adjust font size of tick labels
ax.tick_params(axis='x', labelsize=8)
ax.tick_params(axis='y', labelsize=8)
ax.tick_params(axis='z', labelsize=8)

ax.set_title('Booth Function 3D Surface Plot')
ax.legend()
ax.set_box_aspect(aspect=None, zoom=0.8)

# Add a line from the point (1, 3) to the XY plane
ax.plot([highlight_x, highlight_x], [highlight_y, highlight_y], [highlight_z, 0], color='black', linestyle='dashed')

# Adjust layout
plt.tight_layout()

# Save the plot as a PDF file
plt.savefig('../images/combined_booth_plots.pdf', format='pdf')

# Show the plot (optional)
plt.show()
```

## Six-hump camelback function

**Equation**

$f(x,y)=(4 - 2.1x^2 + \frac{x^4}{3})x^2 + xy + (-4 + 4y^2)y^2$

**Global optimum(s)**

 - $f(0.0898, -0.7126)=-1.0316$
 - $f(-0.0898,07126)=-1.0316$

**Constraints**

$x ∈ [-2, 2], y ∈ [-1, 1]$


```python
def obj_shc(x, y):
    z = (4 - 2.1 * pow(x, 2) + (pow(x, 4) / 3)) * pow(x, 2) + x * y + (-1*4 + 4 * pow(y, 2)) * pow(y, 2)
    return z
```

```python
# Generate data points for visualization
x_values = np.linspace(-2, 2, 100)
y_values = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x_values, y_values)
Z = obj_shc(X, Y)

# Highlight the point (0.0898, -0.7126) with a red dot
highlight_x = 0.0898
highlight_y = -0.7126
highlight_z = obj_shc(highlight_x, highlight_y)

# Create a 2D and 3D subplot side by side
fig = plt.figure(figsize=(10,6))

# Plot the 2D contour plot
ax = fig.add_subplot(1,2,1)
contour = ax.contourf(X, Y, Z, cmap='viridis', levels=30)
ax.scatter([highlight_x], [highlight_y], color='red', label='f(-0.7126)=-1.0316')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Booth Function Contour Plot with Shading')
ax.legend()
fig.colorbar(contour, ax=ax)

# # Plot the 3D surface plot with only 3D axes
ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.scatter([highlight_x], [highlight_y], [highlight_z], color='red', label='f(-0.7126)=-1.0316')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Six-hump camelback Function Value')

# Adjust font size of tick labels
ax.tick_params(axis='x', labelsize=8)
ax.tick_params(axis='y', labelsize=8)
ax.tick_params(axis='z', labelsize=8)

ax.set_title('Six-hump camelback Function 3D Surface Plot')
ax.legend()
ax.set_box_aspect(aspect=None, zoom=0.8)

# Add a line from the point (1, 3) to the XY plane
ax.plot([highlight_x, highlight_x], [highlight_y, highlight_y], [highlight_z, 0], color='black', linestyle='dashed')

# Adjust layout
plt.tight_layout()

# Save the plot as a PDF file
plt.savefig('../images/combined_shc_plots.pdf', format='pdf')

# Show the plot (optional)
plt.show()
```
